Web scraping, the process of extracting data from websites, has emerged as a powerful technique to gather information from the vast expanse of the internet. 

Requests Module
The requests library is used for making HTTP requests to a specific URL and returns the response. Python requests provide inbuilt functionalities for managing both the request and response.

HTML पार्स करणे म्हणजे HTML कोडचे तुकडे करून त्यातील टॅग्स, माहिती आणि रचना ओळखणे, ज्यामुळे त्या माहितीवर पुढील प्रक्रिया करता येते.

r.content:
Type: Returns the content as bytes (raw binary data).
Usage: Used when you need the exact raw data from the response, such as images, PDF files, or other non-text content.

r.text:
Type: Returns the content as a Unicode string (str).
Usage: Used when you want the text representation of the content, like HTML, JSON, or plain text.
Encoding: Automatically decodes the response content based on the response's Content-Type header and the charset specified.

```python
import requests
# Making a GET request
r = requests.get('https://amazon.com')
# check status code for response received
# success code - 200
#print(r)
# print content of request
print(r.content)
#print(r.content) वापरून तुम्ही एखाद्या वेब पेज किंवा API कडून मिळालेल्या कच्च्या डेटा (जसा मिळालाय तसा) ला कन्सोलमध्ये दाखवू शकता.
```

```python
import requests
from bs4 import BeautifulSoup

# Making a GET request
r = requests.get('https://www.geeksforgeeks.org/python-programming-language/')

# check status code for response received
# success code - 200
print(r)

# Parsing the HTML
soup = BeautifulSoup(r.content, 'html.parser')
print(soup.prettify())                         # use for beautify
```

requests हे पॅकेज HTTP विनंती पाठवण्यासाठी वापरले जाते.
BeautifulSoup हे HTML आणि XML दस्तऐवजांचे पार्सिंग करण्यासाठी वापरले जाते.

GET विनंती:
requests.get() फंक्शन वापरून दिलेल्या URL (वेबसाइट) वर GET विनंती पाठवली जाते.
r या वेरिएबलमध्ये HTTP प्रतिसाद (response) साठवला जातो.

प्रतिसादाची तपासणी:
print(r) या वाक्यात, r चा तपशील प्रिंट केला जातो, ज्यामध्ये HTTP स्टेटस कोड समाविष्ट असतो.
200 स्टेटस कोड म्हणजे विनंती यशस्वी झाली आहे.

HTML पार्सिंग:
r.content मध्ये मिळालेल्या HTML डेटा वापरून BeautifulSoup च्या सहाय्याने त्या डेटा चे पार्सिंग केले जाते.
soup या वेरिएबलमध्ये पार्स केलेला HTML डेटा साठवला जातो.

सौंदर्यात्मक प्रदर्शनी:
soup.prettify() फंक्शन वापरून HTML डेटा अधिक वाचनिय स्वरूपात प्रिंट केला जातो. म्हणजे, HTML तागांचे स्तर (indentation) योग्यरित्या दर्शवले जातात, ज्यामुळे त्याचा तपशील लक्षात घेणे सोपे होते.

```python
import requests
from bs4 import BeautifulSoup
response = requests.get('https://en.wikipedia.org/wiki/Python_(programming_language)')
soup = BeautifulSoup(response.text, 'html.parser')
print(soup)
paragraphs = soup.find_all('p')
for paragraph in paragraphs:
    print(paragraph.get_text())       # get_text() हे फंक्शन HTML मधील टॅग्स काढून फक्त त्यामधील मजकूर मिळवण्यासाठी वापरले जाते
```

Here’s a detailed explanation:

response: When you call requests.get(URL), it sends an HTTP request to the specified URL. The server responds with an HTTP response, which is stored in the response object.

response.text: This contains the content of the HTTP response, specifically the body of the web page. It is the HTML source code of the webpage returned by the server, but in string format.

In summary, response.text is the full HTML code of the web page you are trying to scrape, and it is passed to BeautifulSoup to parse and manipulate the HTML.

```python
links = soup.find_all('a')
for link in links:
    print(link.get('href'))
```

स्पष्टीकरण:
links = soup.find_all('a'): ही ओळ HTML मध्ये सर्व <a> (अँकर) टॅग शोधते. <a> टॅग हे वेबपृष्ठावरील हायपरलिंक्स म्हणजेच लिंक असतात. हे सर्व <a> टॅग links नावाच्या यादीत साठवले जातात.

for link in links:: ही एक लूप आहे जी links मधील प्रत्येक लिंकवर (प्रत्येक <a> टॅगवर) जाईल.

print(link.get('href')): प्रत्येक लिंकसाठी, href अॅट्रिब्युटमधील URL काढले जाते, जो त्या लिंकने कोणत्या पृष्ठाकडे नेले जाते हे दर्शवतो. get('href') वापरून तो href अॅट्रिब्युटचा मूल्य (URL) प्रिंट केला जातो

HTML मध्ये, href म्हणजे Hypertext Reference. हे एक अॅट्रिब्युट (गुणधर्म) आहे जो <a> (अँकर) टॅगमध्ये वापरला जातो आणि तो त्या लिंकचे URL (वेबसाईटचे पत्ते) दर्शवतो ज्या पृष्ठावर ती लिंक नेते.
